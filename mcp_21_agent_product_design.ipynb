# MCP 第 21 课：Agent Product Design，强系统不等于强产品

走到这里，已经不能只用“模型强不强”来理解 agent 了。

真正决定用户体验的，常常是产品设计问题：

- 用户如何下达任务
- 系统如何反馈进度
- 失败如何解释
- 工具权限如何表达


## 1. 产品视角和模型视角不一样

模型视角会问：

- 这个 prompt 怎么写
- 这个 tool schema 怎么设计

产品视角会问：

- 用户为什么信任这个系统
- 用户什么时候会中途放弃
- 用户能不能理解系统刚才做了什么


## 2. 一个成熟 agent 产品通常要回答的 5 个问题

- 它能做什么
- 它不能做什么
- 它正在做什么
- 它为什么这么做
- 它失败时怎么办

## 3. 真正重要的体验细节

很多 agent 产品差距，来自这些小地方：

- 任务输入方式是否自然
- 中间进度是否清楚
- 工具使用是否透明
- 确认点是否合理
- 最终报告是否能让人快速判断结果


## 4. 创新不只来自模型

如果你未来想做创新，很值得看的方向包括：

- 任务入口设计
- 多轮协作体验
- 批量任务体验
- trace / diff / evidence 的呈现方式
- 权限和确认的交互方式


## 本课工程直觉

1. **agent 系统做得出来，不等于 agent 产品做得好用。**
2. **产品设计会直接影响信任、效率和留存。**
3. **创新很多时候发生在人机交互层，而不是只发生在模型层。**
